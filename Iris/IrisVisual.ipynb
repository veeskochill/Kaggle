```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
kdata = pd.read_csv('iris.csv', header=0)
```

```python
kdata.columns
```

```python
from itertools import cycle
cycol = cycle('bgrcmk').next

for species, leaf in kdata.groupby('Species'):
    plt.scatter(x = leaf.SepalLengthCm, y = leaf.SepalWidthCm, label = species, color = cycol())

plt.title('Iris Sepal : Length vs Width')
plt.xlabel('Sepal Length cm')
plt.ylabel('Sepal Width cm')
plt.legend()
plt.grid(True)
plt.show()
```
